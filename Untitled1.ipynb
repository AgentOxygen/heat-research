```python
import pop_weighted as pw
import xarray
import numpy as np
```

```python
variable = "HWF"
exp_num = "3136"
min_max = "tx"
weighted, shifted_heat, error = pw.weighted(variable, exp_num, min_max)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from paths import heat_out_trefht_tmax_members_1920_1950_CONTROL as tmax_paths

member_paths = [path for path in tmax_paths()[0] if exp_num in path]
heat_data = xarray.open_mfdataset(member_paths, concat_dim="member", combine="nested")[f"{variable}_{min_max}90"].dt.days.mean(dim="member")

fig, ax1 = plt.subplots(figsize=(20, 10), facecolor='w')
fig.suptitle(f"2020 Population Weighted vs Unweighted {variable} EXP-{exp_num}", fontsize=20)

weighted.mean("lat").mean("lon").plot(ax=ax1, label="Weighted")

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

heat_data.mean("lat").mean("lon").plot(ax=ax2, label="Unweighted", color="red")
ax1.legend(loc='upper left', bbox_to_anchor=(0.05, 1))
ax2.legend(loc='upper left', bbox_to_anchor=(0.05, 0.9))

fig.tight_layout()  # otherwise the right y-label is slightly clipped
fig.savefig("out.png")
plt.show()
```
