```python
from paths import trefhtmx_members as max_paths, trefhtmn_members as min_paths, heat_out_trefht_tmin_members_1920_1950_CONTROL as tmin_paths, heat_out_trefht_tmax_members_1920_1950_CONTROL as tmax_paths
import xarray
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
from matplotlib import colors
from matplotlib import rc
from regionmask.defined_regions import ar6
import dask

variable="HWF"
exp_num="3136"

print("Loading temps")

with dask.config.set(**{'array.slicing.split_large_chunks': False}):
    all_tmax_paths = [path for path in max_paths()[0]]
    xghg_tmax_paths = [path for path in max_paths()[1]]
    xaer_tmax_paths = [path for path in max_paths()[2]]
    
    all_tmax_ds = xarray.open_mfdataset(all_tmax_paths, concat_dim="member", combine="nested")["TREFHTMX"].groupby("time.year").mean()
    xaer_tmax_ds = xarray.open_mfdataset(all_tmax_paths, concat_dim="member", combine="nested")["TREFHTMX"].groupby("time.year").mean()
    xghg_tmax_ds = xarray.open_mfdataset(all_tmax_paths, concat_dim="member", combine="nested")["TREFHTMX"].groupby("time.year").mean()
    
    all_tmin_paths = [path for path in min_paths()[0]]
    xghg_tmin_paths = [path for path in min_paths()[1]]
    xaer_tmin_paths = [path for path in min_paths()[2]]
    
    all_tmin_ds = xarray.open_mfdataset(all_tmin_paths, concat_dim="member", combine="nested")["TREFHTMN"].groupby("time.year").mean()
    xaer_tmin_ds = xarray.open_mfdataset(xaer_tmin_paths, concat_dim="member", combine="nested")["TREFHTMN"].groupby("time.year").mean()
    xghg_tmin_ds = xarray.open_mfdataset(xghg_tmin_paths, concat_dim="member", combine="nested")["TREFHTMN"].groupby("time.year").mean()

print("Temps loaded")
    
all_member_paths = [path for path in tmax_paths()[0] if exp_num in path]
xghg_member_paths = [path for path in tmax_paths()[1] if exp_num in path]
xaer_member_paths = [path for path in tmax_paths()[2] if exp_num in path]

tx_all_dataset = xarray.open_mfdataset(all_member_paths, concat_dim="member", combine="nested")[f"{variable}_tx90"].dt.days / all_tmax_ds
tx_xaer_dataset = xarray.open_mfdataset(xaer_member_paths, concat_dim="member", combine="nested")[f"{variable}_tx90"].dt.days / xaer_tmax_ds
tx_xghg_dataset = xarray.open_mfdataset(xghg_member_paths, concat_dim="member", combine="nested")[f"{variable}_tx90"].dt.days / xghg_tmax_ds

tx_all_data = tx_all_dataset.mean(dim="member")
tx_xaer_data = tx_xaer_dataset.mean(dim="member")
tx_xghg_data = tx_xghg_dataset.mean(dim="member")
land_mask = ar6.land.mask(tx_all_data)

tx_all_data = tx_all_data.where(land_mask > 0).mean(dim="lat").mean(dim="lon")
tx_xaer_data = tx_xaer_data.where(land_mask > 0).mean(dim="lat").mean(dim="lon")
tx_xghg_data = tx_xghg_data.where(land_mask > 0).mean(dim="lat").mean(dim="lon")

all_member_paths = [path for path in tmin_paths()[0] if exp_num in path]
xghg_member_paths = [path for path in tmin_paths()[1] if exp_num in path]
xaer_member_paths = [path for path in tmin_paths()[2] if exp_num in path]

tn_all_dataset = xarray.open_mfdataset(all_member_paths, concat_dim="member", combine="nested")[f"{variable}_tx90"].dt.days / all_tmin_ds
tn_xaer_dataset = xarray.open_mfdataset(xaer_member_paths, concat_dim="member", combine="nested")[f"{variable}_tx90"].dt.days / xaer_tmin_ds
tn_xghg_dataset = xarray.open_mfdataset(xghg_member_paths, concat_dim="member", combine="nested")[f"{variable}_tx90"].dt.days / xghg_tmin_paths
tn_all_data = tn_all_dataset.mean(dim="member")
tn_xaer_data = tn_xaer_dataset.mean(dim="member")
tn_xghg_data = tn_xghg_dataset.mean(dim="member")

tn_all_data = tn_all_data.where(land_mask > 0).mean(dim="lat").mean(dim="lon")
tn_xaer_data = tn_xaer_data.where(land_mask > 0).mean(dim="lat").mean(dim="lon")
tn_xghg_data = tn_xghg_data.where(land_mask > 0).mean(dim="lat").mean(dim="lon")

# f, (ax1, ax2) = plt.subplots(2, 1, figsize=(18, 15), facecolor='w')
# f.suptitle(f"{variable}/{Temperature} {exp_num} Isolated Signal Comparison (Land Masked)", fontsize=30)
# rc('font', **{'weight': 'bold', 'size': 22})
```
